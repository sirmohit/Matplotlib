# SEABORN

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

# Scatter plot

```python
sns.scatterplot(x = iris.sepal_length,y= iris.sepal_width,data = iris,color = 'b')
```

```python
sns.scatterplot(x = iris['sepal_length'],y = iris['petal_length'])
```

```python
sns.scatterplot(x= iris.sepal_length,y= iris.petal_width)
```

# HISTOGRAM PLOT

```python
sns.distplot(iris.petal_length)
```

```python
sns.distplot(iris.sepal_length , kde = False)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x= tips.total_bill,y = tips.tip)
```

```python
tips.smoker.value_counts()
```

# Relational Plot

```python
sns.relplot(x = tips.total_bill,y =tips.tip,data = tips,hue = 'smoker')
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip , data = tips,style = 'size',hue = 'smoker')
```

# Category Plot

```python
sns.catplot(x= 'day',y =  'total_bill',data = tips)
```

```python
sns.catplot(x= 'smoker',y ='tip' ,data = tips)
```

# Joint Plot

```python
sns.jointplot(x= tips.total_bill,y = tips.tip)
```

# Pair plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```
