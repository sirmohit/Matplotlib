```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,200)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x= np.linspace(0,10,200)
y= np.sin(x)
plt.plot(x,y,c= 'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('It is my plot')
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
color = np.random.rand(50)
size =50* np.random.rand(50)
plt.figure(figsize = (15,15))
plt.scatter(x,y,c=color,s= size)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
color = np.random.rand(50)
size = 1000*np.random.rand(50)
plt.figure(figsize = (10,10))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('My plot')
plt.scatter(x,y,c= color,s = size, alpha = .5 , marker = 'v')
plt.show()
```

```python
# setting specific color for each datapoint
x= np.array([1,2,3,4,5])
y= np.array([6,7,8,9,10])
color = ['red','blue' , "#4E301D" , "#DFC010" , "#EC801C" ]
plt.scatter(x,y,c= color,marker = 'v')
plt.show()
```

```python
# Bar Plot
x= ['a','b','c','d','e']
y = [12,23,34,45,56]
plt.bar(x,y,color = "red")
plt.show()
```

```python
# HISTOGRAM PLOT
x=[1,2,3,4,5,6,7,6,5,4,3,23,2,2,1,1]
plt.hist(x)
plt.show()
```

```python
# 3D plot
x= np.random.rand(50)
y= np.random.rand(50)
z= np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.scatter(x,y,z)
plt.show()
```
